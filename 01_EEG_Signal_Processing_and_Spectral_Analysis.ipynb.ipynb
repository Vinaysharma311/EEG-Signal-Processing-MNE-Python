## Dataset

The analysis is performed using the official MNE Sample Dataset. Since the dataset contains both MEG and EEG recordings, only the EEG channels are selected for this project.

```python


data_path = (
    Path("eeg_montage_data")
    / "eeg_montage_data"
    / "MNE-sample-data"
    / "MEG"
    / "sample"
)

raw_file = data_path / "sample_audvis_raw.fif"

raw = mne.io.read_raw_fif(
    raw_file,
    preload=True
)
```

```python
print(raw)
```

```python
raw.info
```

```python
print(raw.ch_names)
```

```python
raw_eeg = raw.copy().pick("eeg")

print(raw_eeg)
print(raw_eeg.ch_names)
```

## Visualizing Raw EEG

The raw EEG signals were visualized to inspect the recording quality and observe the signal before applying any preprocessing techniques.

```python
raw_eeg.plot(
    duration=10,
    n_channels=20,
    scalings="auto",
    block=True
)
```

```python
raw_filtered = raw_eeg.copy()
```

## EEG Preprocessing

A 1–40 Hz band-pass filter was applied to remove low-frequency drift and high-frequency noise while preserving the major EEG frequency bands used for analysis.

```python
raw_filtered.filter(
    l_freq=1.0,
    h_freq=40.0
)
```

```python
raw_eeg.plot(
    duration=10,
    n_channels=20
)
raw_filtered.plot(
    duration=10,
    n_channels=20
)
```

## Spectral Analysis

Power Spectral Density (PSD) was computed to study how the signal power is distributed across different frequencies. This helps identify the dominant EEG frequency bands.

```python
psd = raw_filtered.compute_psd(
    fmin=1,
    fmax=40
)
```

```python
psd.plot()
```

```python
psd.plot(
    average=True
)
```

```python
psd = raw_filtered.compute_psd(
    fmin=1,
    fmax=40
)

psd_data = psd.get_data()
freqs = psd.freqs

print(psd_data.shape)
print(freqs.shape)
```

```python
import numpy as np

def band_power(psd_data, freqs, fmin, fmax):
    idx = np.logical_and(freqs >= fmin, freqs <= fmax)
    return psd_data[:, idx].mean()
```

```python
delta = band_power(psd_data, freqs, 1, 4)
theta = band_power(psd_data, freqs, 4, 8)
alpha = band_power(psd_data, freqs, 8, 13)
beta  = band_power(psd_data, freqs, 13, 30)

print("Delta :", delta)
print("Theta :", theta)
print("Alpha :", alpha)
print("Beta  :", beta)
```

## Brain Wave Analysis

The average power of the Delta, Theta, Alpha, and Beta frequency bands was calculated to compare their contribution in the EEG recording.

```python
import pandas as pd

bands = pd.DataFrame({
    "Band": ["Delta", "Theta", "Alpha", "Beta"],
    "Frequency (Hz)": ["1-4", "4-8", "8-13", "13-30"],
    "Mean Power": [delta, theta, alpha, beta]
})

bands
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))

plt.bar(
    bands["Band"],
    bands["Mean Power"]
)

plt.xlabel("Brain Wave")
plt.ylabel("Mean Power")
plt.title("Average EEG Band Power")

plt.grid(axis="y")

plt.show()
```
